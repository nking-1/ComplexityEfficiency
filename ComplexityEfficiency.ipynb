````python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
c = 3.0e8        # Speed of light (m/s)
hbar = 1.0545718e-34  # Reduced Planck's constant (J·s)
k_B = 1.380649e-23    # Boltzmann constant (J/K)

# Function for black hole entropy (S_BH)
def entropy_black_hole(M):
    """Calculate the entropy of a black hole for a given mass M."""
    return k_B * (4 * np.pi * G * M**2) / (hbar * c)

# Function for rate of entropy change (dS/dt)
def rate_entropy_change(M):
    """Calculate the rate of entropy change (dS/dt) via Hawking radiation."""
    return 1 / M  # Proportional to 1/M (simplified scaling)

# Function for maximum information flow (I_max)
def max_information_flow(M):
    """Calculate the maximum information flow for a black hole."""
    S_BH = entropy_black_hole(M)
    dS_dt = rate_entropy_change(M)
    return S_BH * dS_dt

# Mass range (in kg) - from small black holes to large ones
M_values = np.logspace(22, 30, 500)  # Black hole mass range (kg)

# Calculate values
S_BH_values = entropy_black_hole(M_values)
dS_dt_values = rate_entropy_change(M_values)
I_max_values = max_information_flow(M_values)

# Plot results
plt.figure(figsize=(12, 8))

# Plot entropy
plt.subplot(3, 1, 1)
plt.plot(M_values, S_BH_values, label='Entropy (S_BH)', color='blue')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Black Hole Mass (kg)')
plt.ylabel('Entropy (J/K)')
plt.title('Black Hole Entropy')
plt.grid(True)
plt.legend()

# Plot rate of entropy change
plt.subplot(3, 1, 2)
plt.plot(M_values, dS_dt_values, label='Rate of Entropy Change (dS/dt)', color='green')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Black Hole Mass (kg)')
plt.ylabel('Rate of Entropy Change')
plt.title('Entropy Change Rate')
plt.grid(True)
plt.legend()

# Plot maximum information flow
plt.subplot(3, 1, 3)
plt.plot(M_values, I_max_values, label='Maximum Information Flow (I_max)', color='red')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Black Hole Mass (kg)')
plt.ylabel('Information Flow')
plt.title('Maximum Information Flow')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
c = 3.0e8        # Speed of light (m/s)
hbar = 1.0545718e-34  # Reduced Planck's constant (J·s)
k_B = 1.380649e-23    # Boltzmann constant (J/K)

# Scaling laws fitting function
def power_law(x, a, b):
    return a * x**b

# Black Hole: Scaling Laws Validation
def entropy_black_hole(M):
    return k_B * (4 * np.pi * G * M**2) / (hbar * c)

def rate_entropy_change(M):
    return 1 / M  # Simplified scaling

def max_information_flow(M):
    S_BH = entropy_black_hole(M)
    dS_dt = rate_entropy_change(M)
    return S_BH * dS_dt

# Mass range (in kg)
M_values = np.logspace(15, 40, 1000)  # Broader mass range
S_BH_values = entropy_black_hole(M_values)
dS_dt_values = rate_entropy_change(M_values)
I_max_values = max_information_flow(M_values)

# Fit scaling laws
weights = 1 / S_BH_values
entropy_params, _ = curve_fit(
    power_law, M_values, S_BH_values, sigma=weights, p0=[1, 2], maxfev=10000
)
weights = 1 / dS_dt_values
entropy_rate_params, _ = curve_fit(
    power_law, M_values, dS_dt_values, sigma=weights, p0=[1, -1], maxfev=10000
)

info_flow_params, _ = curve_fit(power_law, M_values, I_max_values)

# Plot Black Hole Scaling Laws
plt.figure(figsize=(10, 6))
plt.loglog(M_values, S_BH_values, label='Entropy (S_BH)', color='blue')
plt.loglog(M_values, power_law(M_values, *entropy_params), '--', color='blue', label=f'Fit: M^{entropy_params[1]:.2f}')
plt.loglog(M_values, dS_dt_values, label='Rate of Entropy Change (dS/dt)', color='green')
plt.loglog(M_values, power_law(M_values, *entropy_rate_params), '--', color='green', label=f'Fit: M^{entropy_rate_params[1]:.2f}')
plt.loglog(M_values, I_max_values, label='Max Info Flow (I_max)', color='red')
plt.loglog(M_values, power_law(M_values, *info_flow_params), '--', color='red', label=f'Fit: M^{info_flow_params[1]:.2f}')
plt.xlabel('Black Hole Mass (kg)')
plt.ylabel('Values (Log Scale)')
plt.title('Scaling Laws for Black Hole Systems')
plt.legend()
plt.grid(True)
plt.show()

plt.loglog(M_values, S_BH_values, label='Entropy (Data)', color='blue')
plt.loglog(M_values, power_law(M_values, *entropy_params), '--', label='Fit', color='red')
plt.legend()
plt.show()


M_normalized = M_values / np.max(M_values)
S_BH_normalized = S_BH_values / np.max(S_BH_values)
entropy_params, _ = curve_fit(
    power_law, M_normalized, S_BH_normalized, p0=[1, 2], maxfev=10000
)
plt.loglog(M_values, S_BH_values, label='Entropy (S_BH)', color='blue')
plt.loglog(M_values, power_law(M_values, *entropy_params), '--', color='blue', label=f'Fit: M^{entropy_params[1]:.2f}')


# Cosmology: Scaling Laws Validation
def cosmological_entropy(R):
    return k_B * R**2

def cosmological_entropy_rate(R):
    rho = 1e-27  # Approximate energy density (kg/m^3)
    H = c / R    # Simplified Hubble parameter
    return R**3 * rho * H

def cosmological_information_flow(R):
    S_cosmo = cosmological_entropy(R)
    dS_dt = cosmological_entropy_rate(R)
    return S_cosmo * dS_dt

# Radius range (meters)
R_values = np.logspace(23, 26, 500)
S_cosmo_values = cosmological_entropy(R_values)
dS_dt_cosmo_values = cosmological_entropy_rate(R_values)
I_max_cosmo_values = cosmological_information_flow(R_values)

# Fit scaling laws with initial guesses and increased iterations
cosmo_entropy_params, _ = curve_fit(
    power_law,
    R_values,
    S_cosmo_values,
    p0=[1, 2],  # Initial guess: amplitude = 1, exponent = 2
    maxfev=10000  # Increase maximum function evaluations
)

cosmo_entropy_rate_params, _ = curve_fit(
    power_law,
    R_values,
    dS_dt_cosmo_values,
    p0=[1, 3],  # Initial guess: amplitude = 1, exponent = 3
    maxfev=10000
)

cosmo_info_flow_params, _ = curve_fit(
    power_law,
    R_values,
    I_max_cosmo_values,
    p0=[1, 5],  # Initial guess: amplitude = 1, exponent = 5
    maxfev=10000
)


# Plot Cosmological Scaling Laws
plt.figure(figsize=(10, 6))
plt.loglog(R_values, S_cosmo_values, label='Entropy (S_cosmo)', color='blue')
plt.loglog(R_values, power_law(R_values, *cosmo_entropy_params), '--', color='blue', label=f'Fit: R^{cosmo_entropy_params[1]:.2f}')
plt.loglog(R_values, dS_dt_cosmo_values, label='Rate of Entropy Change (dS/dt)', color='green')
plt.loglog(R_values, power_law(R_values, *cosmo_entropy_rate_params), '--', color='green', label=f'Fit: R^{cosmo_entropy_rate_params[1]:.2f}')
plt.loglog(R_values, I_max_cosmo_values, label='Max Info Flow (I_max)', color='red')
plt.loglog(R_values, power_law(R_values, *cosmo_info_flow_params), '--', color='red', label=f'Fit: R^{cosmo_info_flow_params[1]:.2f}')
plt.xlabel('Hubble Radius (m)')
plt.ylabel('Values (Log Scale)')
plt.title('Scaling Laws for Cosmological Systems')
plt.legend()
plt.grid(True)
plt.show()

# Quantum Systems: Scaling Laws Validation
def quantum_entropy(dim):
    return k_B * np.log(dim)

def quantum_entropy_rate(E):
    return E / hbar

def quantum_information_flow(dim, E):
    S_quantum = quantum_entropy(dim)
    dS_dt = quantum_entropy_rate(E)
    return S_quantum * dS_dt

# Hilbert space dimensions and energy
dim_values = np.logspace(1, 20, 500)
E_value = 1e-15  # Fixed energy for demonstration
S_quantum_values = quantum_entropy(dim_values)
dS_dt_quantum_values = quantum_entropy_rate(E_value) * np.ones_like(dim_values)
I_max_quantum_values = quantum_information_flow(dim_values, E_value)

# Fit scaling laws
quantum_entropy_params, _ = curve_fit(power_law, dim_values, S_quantum_values)
quantum_info_flow_params, _ = curve_fit(power_law, dim_values, I_max_quantum_values)

# Plot Quantum Scaling Laws
plt.figure(figsize=(10, 6))
plt.loglog(dim_values, S_quantum_values, label='Entropy (S_quantum)', color='blue')
plt.loglog(dim_values, power_law(dim_values, *quantum_entropy_params), '--', color='blue', label=f'Fit: dim^{quantum_entropy_params[1]:.2f}')
plt.loglog(dim_values, I_max_quantum_values, label='Max Info Flow (I_max)', color='red')
plt.loglog(dim_values, power_law(dim_values, *quantum_info_flow_params), '--', color='red', label=f'Fit: dim^{quantum_info_flow_params[1]:.2f}')
plt.xlabel('Hilbert Space Dimension')
plt.ylabel('Values (Log Scale)')
plt.title('Scaling Laws for Quantum Systems')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Parameters
dim_values = np.logspace(1, 20, 500)  # Hilbert space dimensions
E_value = 1e-15  # Fixed energy in Joules

# Calculate quantum values
S_quantum_values = quantum_entropy(dim_values)
dS_dt_quantum_values = quantum_entropy_rate(E_value) * np.ones_like(dim_values)
I_max_quantum_values = quantum_information_flow(dim_values, E_value)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11       # Gravitational constant (m^3/kg/s^2)
c = 3e8               # Speed of light (m/s)
hbar = 1.0545718e-34  # Reduced Planck constant (J·s)
k_B = 1.380649e-23    # Boltzmann constant (J/K)
l_p = np.sqrt(hbar * G / c**3)  # Planck length (m)

# Precomputed constants for optimization
hawking_constant = hbar * c**6 / (G**2)  # Hawking radiation power factor
schwarzschild_factor = 2 * G / c**2      # Schwarzschild radius factor

# Black hole parameters (scaled down for faster dynamics)
M = 1e15  # Initial black hole mass (kg)
R_s = schwarzschild_factor * M  # Schwarzschild radius (m)

# Time and step parameters
t_max = 1e8           # Extended simulation time to observe dynamics (s)
dt = 1e2              # Refined time step
steps = int(t_max / dt)

# Arrays for tracking values
time = np.zeros(steps)
R_s_values = np.zeros(steps)
M_values = np.zeros(steps)
I_max_values = np.zeros(steps)
dI_max_values = np.zeros(steps)  # Rate of change of I_max

# Initial conditions
R_s_values[0] = R_s
M_values[0] = M

# Simulation loop
for n in range(steps - 1):
    # Compute mass loss rate
    dot_M = max(-hawking_constant / M_values[n]**2, -1e20)  # Capped mass loss rate
    M_values[n + 1] = max(M_values[n] + dot_M * dt, 1e-40)  # Avoid zero mass

    # Update Schwarzschild radius
    R_s_values[n + 1] = max(schwarzschild_factor * M_values[n + 1], l_p)  # Regularized at Planck scale

    # Compute maximum information flow
    rho = M_values[n + 1] / (R_s_values[n + 1]**3 + l_p**3)  # Regularized density
    I_max_values[n] = max((k_B**2 * rho**2 * R_s_values[n + 1]**3 * c) / G, 1e-12)  # Avoid divergence

    # Compute rate of change of I_max
    if n > 0:
        dI_max_values[n] = (I_max_values[n] - I_max_values[n - 1]) / dt

    # Print debugging information every 10,000 steps
    if n % 10000 == 0:
        print(f"Step {n}: M = {M_values[n]:.3e}, R_s = {R_s_values[n]:.3e}, I_max = {I_max_values[n]:.3e}")

    # Stop if mass reaches the Planck scale
    if M_values[n + 1] < 1e-8:
        print(f"Black hole mass reached Planck scale at step {n}")
        break

# Filter valid values for plotting
valid_indices = (M_values > 0) & (R_s_values > 0) & (I_max_values > 0)
time_valid = time[valid_indices]
M_valid = M_values[valid_indices]
R_s_valid = R_s_values[valid_indices]
I_max_valid = I_max_values[valid_indices]
dI_max_valid = dI_max_values[valid_indices]

# Plotting
plt.figure(figsize=(12, 8))

# Mass, Radius, and Information Flow
plt.subplot(3, 1, 1)
plt.plot(time_valid, M_valid, label="Black Hole Mass (M)")
plt.plot(time_valid, R_s_valid, label="Schwarzschild Radius (R_s)")
plt.plot(time_valid, I_max_valid, label="I_max (Information Flow)")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.title("Black Hole Dynamics")
plt.legend()
plt.grid(True)

# Rate of Change of I_max
plt.subplot(3, 1, 2)
plt.plot(time_valid, dI_max_valid, label="dI_max/dt (Rate of Change of I_max)", color="orange")
plt.xscale("log")
plt.xlabel("Time (s)")
plt.ylabel("Rate of Change")
plt.title("Rate of Change of I_max")
plt.legend()
plt.grid(True)

# Zoomed-in Transition Phase
plt.subplot(3, 1, 3)
transition_indices = (time_valid > 1e7) & (time_valid < 1e8)
plt.plot(time_valid[transition_indices], I_max_valid[transition_indices], label="I_max (Zoomed)", color="green")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Time (s)")
plt.ylabel("I_max")
plt.title("Zoomed View of Transition Phase")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

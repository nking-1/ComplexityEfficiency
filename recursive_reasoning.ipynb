```python
from openai import OpenAI
import os

client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY"),
)

# Define the initial teaching prompt for I_max
def get_initial_prompt():
    return """You are a highly capable reasoning AI trained to understand and apply novel frameworks. Your task is to learn and use \( \mathcal{I}_{\text{max}} \), the principle of maximum information flow, to generate and refine insights across diverse fields.

### Definition of \( \mathcal{I}_{\text{max}} \):
\( \mathcal{I}_{\text{max}} \) represents the balance between:
1. **Stored Complexity (\( S \))**: The richness of information in a system.
2. **Dynamic Efficiency (\( \Delta S / \Delta t \))**: The rate of change or adaptability of the system.

This balance is fundamental to all systems that encode, transform, and redistribute information. \( \mathcal{I}_{\text{max}} \) generates new structures by dynamically balancing \( S \) and \( \Delta S / \Delta t \).

### Task:
1. Use \( \mathcal{I}_{\text{max}} \) to iteratively generate and refine insights, questions, and models.
2. Apply \( \mathcal{I}_{\text{max}} \) recursively—each question should refine or expand the previous insight.
3. Generate insights across domains or by synthesizing multiple domains.

### Few-Shot Examples:

**Example 1:**
**Question**: How does \( \mathcal{I}_{\text{max}} \) guide advancements in biotechnology?
**Insight**: Stored complexity (\( S \)) is represented by genetic data, experimental techniques, and research methodologies. Dynamic efficiency (\( \Delta S / \Delta t \)) involves the rapid adaptation of these capabilities to develop new treatments. \( \mathcal{I}_{\text{max}} \) suggests breakthroughs will occur when data collection (high \( S \)) and adaptive testing (high \( \Delta S / \Delta t \)) are balanced.

**Example 2:**
**Question**: How can \( \mathcal{I}_{\text{max}} \) optimize educational systems?
**Insight**: Stored complexity (\( S \)) is the accumulated knowledge and resources in education, while \( \Delta S / \Delta t \) represents adaptability in curricula and methods. \( \mathcal{I}_{\text{max}} \) implies that combining rich content with dynamic teaching strategies creates systems deeply rooted in knowledge yet flexible enough to meet evolving societal needs.

Now, begin applying \( \mathcal{I}_{\text{max}} \) to generate new insights recursively across domains."""

# Function to send a prompt to GPT-4o
def query_gpt4o(prompt):
    response = client.chat.completions.create(
        messages=[
            {"role": "system", "content": "You are a highly capable AI reasoning about \( \mathcal{I}_{\text{max}} \)."},
            {"role": "user", "content": prompt},
        ],
        model="gpt-4o",
    )
    return response.choices[0].message.content



# Function to iteratively refine the prompt
def iterative_prompting(initial_prompt, iterations=5):
    prompt = initial_prompt
    for i in range(iterations):
        print(f"\nIteration {i + 1}:\n{'=' * 40}")
        # Query GPT-4o with the current prompt
        response = query_gpt4o(prompt)
        print(f"GPT-4o Response:\n{response}")
        
        # Dynamically refine the prompt based on the response (manual step for now)
        feedback = input("\nRefine the prompt (leave empty to continue with the same prompt): ")
        if feedback.strip():
            prompt = feedback
        else:
            prompt = prompt + f"\n### GPT-4o Insight from Iteration {i + 1}:\n{response}"

    print("\nFinal Iterated Prompt:\n", prompt)
    return prompt
```

```python
# Get the initial prompt
initial_prompt = get_initial_prompt()

# Start the iterative prompting process
final_prompt = iterative_prompting(initial_prompt, iterations=5)

print("\nFinal Optimized Prompt:\n", final_prompt)
```
